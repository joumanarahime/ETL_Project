```python
import pandas as pd
from sqlalchemy import create_engine
import csv
import os.path
import os

cwd = os.getcwd()  # Get the current working directory (cwd)
files = os.listdir(cwd)  # Get all the files in that directory
```

```python
csv_file=f"{cwd}/Resources/student_loan_debt.csv"
student_loan_data = pd.read_csv(csv_file)
student_loan_data.head()
student_loan_data.head()
```

```python
student_loan_df = student_loan_data[['Unnamed: 0','Unnamed: 1',
                                     'Unnamed: 7','Unnamed: 8',
                                     'Unnamed: 15','Unnamed: 16',
                                     'Unnamed: 19','Unnamed: 20',
                                     'Unnamed: 33','Unnamed: 34'
                                   ]].copy()

student_loan_df=student_loan_df.rename(columns={'Unnamed: 0':'County',
                                              'Unnamed: 1':'State',
                                              'Unnamed: 7':'Median Loan White',
                                              'Unnamed: 8':'Median Loan Nonwhite',
                                              'Unnamed: 15':'In Collection Loan White',
                                              'Unnamed: 16':'In Collection Loan Nonwhite',
                                              'Unnamed: 19':'Median Monthly Payment White',
                                              'Unnamed: 20':'Median Monthly Payment Nonwhite',
                                              'Unnamed: 33':'Avg Household Income White',
                                              'Unnamed: 34':'Avg Household Income Nonwhite'
                                             })
student_loan_df.head()
```

```python
# drop columns that have NAN
student_loan_df = student_loan_df.dropna()
student_loan_df.head()
```

```python
# drop first row that has the column names 
student_loan_df = student_loan_df.drop([0])
student_loan_df.head()
```

```python
# select all rows that does not have  Median Loans for both White and Nonwhite

select_rows = student_loan_df[student_loan_df['Median Loan White'].isin(['n/a*', 'n/a**']) &
                         student_loan_df['Median Loan Nonwhite'].isin(['n/a*', 'n/a**']) 
                        ]

index_list = select_rows.index.values  

for index in index_list:
    student_loan_df = student_loan_df.drop([index])
student_loan_df

```

```python
# connect to the databse
# rds_connection_string = "<>:<password>@localhost:53577/students_loan_db"
import psycopg2 as pg
import pandas.io.sql as psql

connection_string = "joumana:*****@localhost:5432/student_loan_db"
engine = create_engine(f'postgresql://{connection_string}')
```

```python
student_loan_df.to_sql(name='student_loan', con=engine, if_exists='append', index=False)
```
